```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (11, 7)
```

```python
df =  pd.read_csv('dataFine.csv')
```

```python
data_1_t = pd.pivot_table(df, values="count",index=["year",'month'], aggfunc="sum").reset_index()
```

```python
(sns.lineplot(data=data_1_t,x="year", y= "count",ci=False)).set(title='Динамика числа обращений в службу занятости', xlabel= 'Год', ylabel='Количество обратившихся')
plt.savefig('Task1_1.png')
```

```python
data_1_t_2 = pd.pivot_table(df[df["hc_repeat"]==True], values="count",index=["year",'month'], aggfunc="sum").reset_index()
```

```python
(sns.lineplot(data=data_1_t_2,x="year", y="count",ci=False)).set(title='Динамика числа повторных обращений в органы занятости', xlabel='Год', ylabel='Количество обращений')
```

```python
merged_groups = data_1_t.merge(data_1_t_2, on= ['year', 'month'], how = 'left')
#merged_groups
```

```python
l=sns.lineplot(data = merged_groups, x = 'year', y = 'count_x', ci=False, label = 'Общее количество обращений', marker='o')
(sns.lineplot(data = merged_groups, x = 'year', y = 'count_y', ci=False, label = 'Повторных обращений', marker='o'))
l.set(title='Динамика обращений в службу занятых',xlabel='Год, укрупнённо',ylabel = 'Количество обращений')


temp_data = pd.pivot_table(merged_groups,values = ['count_y', 'count_x'], index = 'year').reset_index()

for x, y in zip(temp_data['year'], temp_data['count_x']):
    plt.text(
        x = x-0.055, 
        y = y-7500, 
        s = '{:.0f}'.format(y),
        color = 'black').set_backgroundcolor('orange')
    
for x, y in zip(temp_data['year'], temp_data['count_y']):
    plt.text(
        x = x-0.055, 
        y = y+4000, 
        s = '{:.0f}'.format(y),
        color = 'white').set_backgroundcolor('#3682ba')

plt.savefig('Task1_1.png')
```

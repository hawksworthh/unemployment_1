```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (11, 7)
```

```python
df =  pd.read_csv('dataFine.csv')
```

```python
df['district'].unique()
```

```python
group_by = pd.pivot_table(df, values="count",index=["year",'month','education','gender', 'hc_didnotwork', 'find_new_job','hc_repeat', 'age', 'district'], aggfunc="sum").reset_index()
```

```python
sns.barplot(data=group_by,x='year',y='count',ci=False,hue='district').set(xlabel='Год, месяц', ylabel = 'Количество подавших повторное заявление', title = 'Динамика повторно подавших заявления')
plt.legend(title='Округ')
plt.savefig('Task9_1.png')
```

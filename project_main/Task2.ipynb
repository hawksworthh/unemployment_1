```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (11, 7)
```

```python
df =  pd.read_csv('dataFine.csv')
```

```python
data_2_t = pd.pivot_table(df, values="count",index=["year",'month','age'], aggfunc="sum").reset_index()
```

```python
#data_2_t
```

```python
month = [ 'Январь', 'Февраль', 'Март', 'Апрель', 'Май', 'Июнь', 'Июль', 'Август', 'Сентябрь', 'Октябрь', 'Ноябрь', 'Декабрь']
data_2_t['month'] = pd.Categorical(data_2_t['month'], categories = month, ordered=True)
```

```python
sns.set_theme()

sns.barplot(data=data_2_t,x='year',y='count', hue = 'age', ci=False).set(xlabel='Год', ylabel = 'Количество обращений за период', title = 'Динамика обращений по возрастным группам')
plt.legend(title='Возрастная группа', loc='upper left')

plt.savefig('Task2_1.png')
```

```python
fig, axes = plt.subplots(2, 2, figsize=(30, 10))


sns.set_theme()

g1 = sns.barplot(ax=axes[0,0], data=data_2_t[data_2_t['year']==2017], x='month',y='count', hue = 'age').set(xlabel='2017 Год', ylabel = 'Количество обращений за период', title = 'Динамика обращений по месяцам')
g2 = sns.barplot(ax=axes[0,1], data=data_2_t[data_2_t['year']==2018], x='month',y='count', hue = 'age').set(xlabel='2018 Год', ylabel = 'Количество обращений за период', title = 'Динамика обращений по месяцам')
g3 = sns.barplot(ax=axes[1,0], data=data_2_t[data_2_t['year']==2019], x='month',y='count', hue = 'age').set(xlabel='2019 Год', ylabel = 'Количество обращений за период', title = 'Динамика обращений по месяцам')
g4 = sns.barplot(ax=axes[1,1], data=data_2_t[data_2_t['year']==2020], x='month',y='count', hue = 'age').set(xlabel='2020 Год', ylabel = 'Количество обращений за период', title = 'Динамика обращений по месяцам')

plt.savefig('Task2_2.png')

plt.show()
```
